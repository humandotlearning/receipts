```
def l_nm(o_nm='shop_name',res=res):
  l=[]
  for a in res:
    #o_nm = a[0]
    #print(a[1])
    if a[0] == o_nm:
      l.append(a[1])
      
  return (o_nm,l)

def dic(res=res,o_nm=o_nm):
  dic={}
  for o in o_nm:
    nm,v =l_nm(o)
    v= {int(x.split('\t')[0].strip('%')): x.split('\t')[1].replace('(','').replace(')','').replace(':','').split(' ') for x in v}
    dic[nm]=v
  return dic


#pp(o_d)
```

```

o_d = dic()

for key,val in o_d.items():
  #pp(val)
  #print(key,val)
  if bool(o_d[key]):
    for k,v in val.items():
      #print(k)
      i=0
      bbox=[]
      for l in v:
        if l=='':
          continue
        #print(l)
        bbox.append(l)

      d= defaultdict(lambda: 0)
      d= {
        'x1':int(bbox[1]) ,                 #xmin
        'y1':int(bbox[3]) ,                 #ymin
        'x2':int(bbox[5]) +  int(bbox[1]),            #xmax
        'y2':int(bbox[7]) +  int(bbox[3])            #ymax 
      }
      o_d[key][k]= d
 
    #o_d.pop(key,None)
    
    
    #print('\n', d)
```

```
def map2(im=im, image=image):
  mp=plt.figure(figsize=(20,20))
  plt.rcParams["axes.grid"] = False
  f,ax =plt.subplots(1,2)
  f.set_figheight(20)
  f.set_figwidth(20)
  ax[0].imshow(im)
  ax[1].imshow(image)

#todo - check if the element exist
def get_croped_image(wh='shop_name',img=img,final=final): 
  '''
  get croped image from the bounding box
  '''
  img_name=img.split('/')[-1]
  #print(final[img_name][wh])
  if wh in final[img_name]:
    b=final[img_name][wh]
    #large= max(k for k, v in b.items())
    for k,v in b.items():
      #bb = b[large]
      bb = b[k]
      image=cv2.imread(img)
      #image=cv2.cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
      image=cv2.cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      #print(bb)
    return image[bb['y1']:bb['y2'], bb['x1']:bb['x2']]
  else:
    return False  
    
  #return image[bb['y1']:bb['y2'], bb['x1']:bb['x2']]
```
